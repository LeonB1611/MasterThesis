```python
def FigsToGif(frame_location):
    '''
    frame_location: PATH to folder with frames of the SVG frames
    '''
    frames = [Image.open(image) for image in glob.glob(f"{frame_location}/*.SVG")] #frames from folder i.e. each gif new folder, sorts by name
    frame_one = frames[0]
    frame_one.save("my_awesome.gif", format="GIF", append_images=frames,
               save_all=True, duration=len(frames)*500, loop=0) #time in ms, here 0.5s per frame
```

```python
def OscGif(framefolder,frameending,savefolder,gifname,dur=1000):
    '''
    
    Inputs (type: str):
    
    framefolder: folder where the frames are stored e.g. "Osc_Param_Test_Plots/PlotSeries2/Th14"
    frameending: ending name of the frames to be specified e.g. "*_Th14_sneu_mu_ini.png"
    
    savefolder: folder where gif will be saved "Osc_Param_Test_Gifs"
    gifname: name of gif output e.eg. "Plotseries2_th14_sneu_mu_ini.gif"
    
    dur: frame duration in ms, default 1000ms
    
    '''
    
    frameloc = framefolder+"/"+frameending #frame location string
    save = savefolder+"/"+gifname #save gif string
    
    frames = []
    imgs = sorted(glob.glob(frameloc)) #works
    for i in imgs:
        new_frame = Image.open(i)
        temp = Image.open(imgs[0]) 
        new_frame = new_frame.resize(temp.size) # resizes all frames to first image size
        frames.append(new_frame)

    # Save into a GIF file that loops forever
    frames[0].save(save, format='GIF',
                   append_images=frames[1:],
                   save_all=True,
                   duration=dur, loop=0)
```
