## Contour Plots

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
csv_path = '/afs/ifh.de/user/b/buchholl/scratch/software/fridge_old/analysis/oscnext_sterile/scripts/plotting/'
```

```python
reference_contours = np.loadtxt(
    csv_path + "old_reference_wilks_contour_90pct.csv", delimiter=",",skiprows=1
)
reproduced_contours = np.loadtxt(
    csv_path + "old_repro_wilks_contour_90pct.csv", delimiter=",",skiprows=1
)
reproduced_no_detsys_contours = np.loadtxt(
    csv_path + "old_repro_no_detsys_wilks_contour_90pct.csv", delimiter=",",skiprows=1
)
flercnn_no_detsys_contours = np.loadtxt(
    csv_path + "flercnn_no_detsys_wilks_contour_90pct.csv", delimiter=",",skiprows=1
)
flercnn_pd_no_detsys_contours = np.loadtxt(
    csv_path + "flercnn_pd_no_detsys_wilks_contour_90pct.csv", delimiter=",",skiprows=1
)
```

```python
plt.figure(figsize=(7,5))
plt.plot(reference_contours[:, 0],reference_contours[:, 1], linestyle = "dotted",color='blue',
        label='Reference Contours from Old Analysis')
plt.plot(reproduced_contours[:, 0],reproduced_contours[:, 1], linestyle = "dashed", dashes=(5,8),color='blue',
        label='Reproduced Contours for Old Analysis')
plt.plot(reproduced_no_detsys_contours[:, 0],reproduced_no_detsys_contours[:, 1], linestyle = "solid",color='blue',
        label='Reproduced Contours for Old Analysis without det. sys.')
plt.plot(flercnn_no_detsys_contours[:, 0],flercnn_no_detsys_contours[:, 1], linestyle = "dashed",color='red',
        label='FLERCNN adjusted PID-binning Contours without det. sys.')
#plt.plot(flercnn_pd_no_detsys_contours[:, 0],flercnn_pd_no_detsys_contours[:, 1], linestyle = "dotted",color='red',
#        label='FLERCNN adjusted PID-binning pseudodata Contours without det. sys.')
legend_title = f"all contours {'90%'} C.L."
plt.legend(
    fontsize=9,
    loc="upper center",
    title=legend_title,
    title_fontsize=10,
)
plt.xlim((1e-3, 0.2))
plt.ylim((3e-3, 1))            
plt.ylabel(r"$|U_{\tau 4}|^2 = \sin^2\theta_{34}\cdot\cos^2\theta_{24}$")
plt.xlabel(r"$|U_{\mu 4}|^2 = \sin^2\theta_{24}$")
plt.yscale('log')
plt.xscale('log')
```
