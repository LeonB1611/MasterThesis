```python
import time
set_verbosity(1)

template_maker.pipelines[0].params["theta14"] = 0 * ureg["degree"]
template_maker.pipelines[0].params["deltacp14"] = 0 * ureg["degree"]
template_maker.pipelines[0].params["deltacp24"] = 0 * ureg["degree"]

filter_scan = np.geomspace(1e-7, 1e-2, 15)
cutoff_dict = {}
time_dict = {}
# for dm41 in [0.01, 0.05, 0.1, 0.5, 1, 5, 10]:
for dm41 in [0.1]:
    chi2_cutoff = []
    times = []
    data_maker.pipelines[0].params["theta24"] = 15 * ureg["degree"]
    data_maker.pipelines[0].params["theta34"] = 0 * ureg["degree"]
    data_maker.pipelines[0].params["deltam41"] = dm41 * ureg["eV**2"]
    ps_map_set = data_maker.get_outputs(return_sum=False)[0]
    print(f"dm41: {dm41}")
    for cutoff in filter_scan:
        template_maker.pipelines[0].stages[3].abs_err = cutoff
        template_maker.pipelines[0].stages[3].rel_err = cutoff
        template_maker.pipelines[0].stages[3].setup()
        template_maker.pipelines[0].params["theta24"] = 15 * ureg["degree"]
        template_maker.pipelines[0].params["theta34"] = 0 * ureg["degree"]
        template_maker.pipelines[0].params["deltam41"] = dm41 * ureg["eV**2"]
        start_t = time.time()
        tmp_map_set = template_maker.get_outputs(return_sum=False)[0]
        eval_time = time.time() - start_t
        times.append(eval_time)
        chi2 = tmp_map_set["numu_cc"].chi2(ps_map_set["numu_cc"])
        print(f"err magnitude: {cutoff:.4g}   chi2: {chi2:.4g}  eval time: {eval_time}")
        chi2_cutoff.append(chi2)
    cutoff_dict[dm41] = chi2_cutoff
    time_dict[dm41] = times
```
